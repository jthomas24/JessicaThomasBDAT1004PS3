```python
#Question 6

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

marriages =pd.read_csv('us-marriages-divorces-1867-2014.csv')
marriages.set_index('Year', inplace = True)
marriages.head()



columns = marriages.columns.drop(['Marriages','Divorces','Population'])

fig,ax = plt.subplots()

ax.plot(marriages.index,marriages[columns])
   

ax.set_title('Marriages and Divorces per Capita')
ax.set_xlabel("Year")
ax.set_ylabel("# of Marriages")
ax.legend(columns)
```

```python
#Question 7


import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

marriages =pd.read_csv('us-marriages-divorces-1867-2014.csv',header=0)
marriages.set_index('Year', inplace=True)
marriages.head()

specific_years = [1900,1950,2000]
columns = marriages.columns.drop(['Marriages','Divorces','Population'])
y_values = marriages.loc[specific_years,columns]

fig, ax = plt.subplots()

y_values.plot.bar(ax=ax)

ax.set_title('Marriages and Divorces per Capita in 1900, 1950 and 2000')
ax.set_xlabel("Year")
ax.set_ylabel("# of Marriages")
```
